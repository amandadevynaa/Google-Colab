# **Preprossesing Dataset**

```python
# File ID dari link Google Drive
file_id = "1tCpNxu8qUdlwHtS4-T6NmYvGsITd5txU"
url = f"https://drive.google.com/uc?id={file_id}"

# Baca CSV langsung dari Google Drive
df = pd.read_csv(url)

output = "flood2.csv"
df.to_csv(output, index=False)
df.head()
```

```python
missing_value = df.isnull().sum()
print(f"Jumlah missing value di seluruh kolom: \n{missing_value}")
```

```python
df.dropna(inplace=True)
print("\nJumlah missing value setelah penanganan:")
print(df.isnull().sum())
print("\nJumlah baris setelah penanganan missing value:", df.shape[0])
```

```python
# Mengecek baris yang duplikat di seluruh kolom
duplicate_rows = df.duplicated()

num_duplicate_rows = duplicate_rows.sum()

print(f"Jumlah baris duplikat di seluruh kolom: {num_duplicate_rows}")
```

```python
X = df.drop(columns=['FloodProbability'])
y = df['FloodProbability']

print("\nFitur (X):")
print(X.head())

print("\nTarget (y):")
print(y.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("\nData Setelah Standarisasi:")
print(X_scaled.head())
```

```python
print("\nUkuran Data Setelah Preprocessing:", X_scaled.shape)
print("Tidak ada missing value lagi:", X_scaled.isnull().sum().sum() == 0)
```
